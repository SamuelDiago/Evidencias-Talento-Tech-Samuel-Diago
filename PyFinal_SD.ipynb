### 2. Limpieza y Conversión de Datos

- La columna `año` será convertida a tipo `object`.
- Las columnas `Produccion_toneladas`, `Latitud` y `Longitud` serán limpiadas reemplazando las comas por puntos y luego convertidas a tipo numérico (`float`).

Verificamos que estas conversiones se realizaron correctamente mostrando la información actualizada del DataFrame y sus estadísticas descriptivas.


```python
print('--- Realizando limpieza de datos ---')

# 1. Convertir la columna 'año' a tipo object (string)
df['año'] = df['año'].astype(str)

# 2. Reemplazar comas por puntos y convertir a numérico en las columnas especificadas
columns_to_clean = ['Produccion_toneladas', 'Latitud', 'Longitud']

for col in columns_to_clean:
    if df[col].dtype == 'object': # Only process if it's an object/string type
        df[col] = df[col].str.replace(',', '.', regex=False)
        # Attempt to convert to numeric, coercing errors to NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')

print('\n--- Información del Dataset después de la limpieza (df.info()) ---')
df.info()

print('\n--- Primeras filas del Dataset después de la limpieza (df.head()) ---')
display(df.head(9999))

print('\n--- Estadísticas Descriptivas (df.describe()) después de la limpieza ---')
display(df.describe())
```

1. Centro   5,471,119.70 Ton - 41.04%

```python
import numpy as np
import pandas as pd

# 1. Obtener todos los años únicos presentes en el dataset
todos_los_anios = set(df['año'].unique())

# 2. Identificar cultivos que tienen producción en todos los años
cultivos_unicos_en_todos_los_anios = df.groupby('Cultivo').filter(
    lambda x: todos_los_anios.issubset(x['año'].unique())
)['Cultivo'].unique()

print("\n--- Producción total por Cultivo y Subregión (solo cultivos con producción en todos los años) ---")

if len(cultivos_unicos_en_todos_los_anios) > 0:
    # Filtrar el DataFrame original para incluir solo los cultivos que tienen producción en todos los años
    df_cultivos_all_years = df[df['Cultivo'].isin(cultivos_unicos_en_todos_los_anios)]

    # Crear la tabla pivote
    tabla_produccion_subregion = pd.pivot_table(
        df_cultivos_all_years,
        index='Cultivo',
        columns='Subregion',
        values='Produccion_toneladas',
        aggfunc='sum',
        margins=True, # Añade totales de filas y columnas
        margins_name='Total General' # Nombre para los totales
    )

    # Reemplazar NaN con 0
    tabla_produccion_subregion = tabla_produccion_subregion.fillna(0)

    # Ordenar la tabla por 'Total General' de mayor a menor
    tabla_produccion_subregion = tabla_produccion_subregion.sort_values(by='Total General', ascending=False)

    # Calculate the grand total production from the 'Total General' row and column
    grand_total_production = tabla_produccion_subregion.loc['Total General', 'Total General']

    # Add the 'Porcentaje Total' column
    if grand_total_production > 0:
        tabla_produccion_subregion['Porcentaje Total'] = (tabla_produccion_subregion['Total General'] / grand_total_production) * 100
    else:
        tabla_produccion_subregion['Porcentaje Total'] = 0.0 # Assign a float 0.0 if grand_total is 0

    # Format the 'Porcentaje Total' column to include '%' sign and two decimal places
    # This will convert the column to object (string) type for display
    tabla_produccion_subregion['Porcentaje Total'] = tabla_produccion_subregion['Porcentaje Total'].map('{:,.2f}%'.format)

    # Identify subregion columns for individual percentage calculations
    subregion_cols = [col for col in tabla_produccion_subregion.columns if col not in ['Total General', 'Porcentaje Total']]

    # Create a new DataFrame to store the updated table with percentages
    tabla_final = tabla_produccion_subregion.copy()

    # Iterate over each subregion column to calculate and add percentage columns
    for subregion in subregion_cols:
        # Get the total production for the current subregion
        total_subregion_produccion = tabla_produccion_subregion.loc['Total General', subregion]

        # Calculate percentages, handling division by zero for subregions with no production
        if total_subregion_produccion > 0:
            tabla_final[f'{subregion} %'] = (tabla_produccion_subregion[subregion] / total_subregion_produccion) * 100
        else:
            tabla_final[f'{subregion} %'] = 0.0

        # Format the percentage column
        tabla_final[f'{subregion} %'] = tabla_final[f'{subregion} %'].map('{:,.2f}%'.format)

    # Reorder columns to interleave original production and new percentage columns
    ordered_columns = []
    for subregion in subregion_cols:
        ordered_columns.append(subregion)
        ordered_columns.append(f'{subregion} %')
    ordered_columns.append('Total General')
    ordered_columns.append('Porcentaje Total')

    # Apply the new column order
    tabla_final = tabla_final[ordered_columns]

    # --- Start of new code for 'Subtotal' row ---
    subtotal_values = []
    for col in tabla_final.columns:
        if col in subregion_cols: # This is a subregion production column (e.g., 'Centro')
            subtotal_values.append(tabla_produccion_subregion.loc['Total General', col])
        elif col.endswith(' %') and col.replace(' %', '') in subregion_cols: # This is a subregion percentage column (e.g., 'Centro %')
            subregion_name = col.replace(' %', '')
            subregion_total_production = tabla_produccion_subregion.loc['Total General', subregion_name]
            if grand_total_production > 0:
                percentage_of_valley_total = (subregion_total_production / grand_total_production) * 100
            else:
                percentage_of_valley_total = 0.0
            subtotal_values.append('{:,.2f}%'.format(percentage_of_valley_total))
        elif col == 'Total General':
            subtotal_values.append(grand_total_production)
        elif col == 'Porcentaje Total':
            subtotal_values.append('100.00%')

    # Create a Series with the subtotal values, and set its name as the index label
    subtotal_row_series = pd.Series(subtotal_values, index=tabla_final.columns, name='Subtotal')

    # Concatenate the existing tabla_final with the new subtotal row
    # The 'Total General' row is currently the last row in tabla_final.
    # We want 'Subtotal' to be *below* 'Total General'.
    tabla_final = pd.concat([tabla_final, subtotal_row_series.to_frame().T])
    # --- End of new code ---

    # Quitar la notación científica y mostrar hasta 2 decimales para otras columnas float
    pd.options.display.float_format = '{:,.2f}'.format
    display(tabla_final)

else:
    print("No hay cultivos para generar la tabla de producción por subregión.")

print("""El Plátano, Piña, Banano y Guayaba estan presentes en todo el valle.
Bananito, Coco, Borojo y Chontaduro son princpales del Pacificio (Buenaventura).
En general los demas cultivos entan presente en el casi todo el Valle""")
```

```python
print('--- Producción total por Subregion y Relieve Predominante ---')

# Crear la tabla pivote con Subregion como índice y Relieve Predominante como columnas
tabla_produccion_subregion_relieve = pd.pivot_table(
    df,
    index='Subregion',
    columns='Relieve Predominante',
    values='Produccion_toneladas',
    aggfunc='sum',
    margins=True, # Añade totales de filas y columnas
    margins_name='Total General' # Nombre para los totales
)

# Reemplazar NaN con 0
tabla_produccion_subregion_relieve = tabla_produccion_subregion_relieve.fillna(0)

# Calcular el gran total de producción para el valle
grand_total_production_relieve = tabla_produccion_subregion_relieve.loc['Total General', 'Total General']

# Añadir la columna 'Porcentaje Total' para las filas
if grand_total_production_relieve > 0:
    tabla_produccion_subregion_relieve['Porcentaje Total'] = (tabla_produccion_subregion_relieve['Total General'] / grand_total_production_relieve) * 100
else:
    tabla_produccion_subregion_relieve['Porcentaje Total'] = 0.0

# Formatear la columna 'Porcentaje Total' a dos decimales y con signo '%'
tabla_produccion_subregion_relieve['Porcentaje Total'] = tabla_produccion_subregion_relieve['Porcentaje Total'].map('{:,.2f}%'.format)

# Identificar las columnas de relieve para cálculos individuales de porcentaje
relieve_cols = [col for col in tabla_produccion_subregion_relieve.columns if col not in ['Total General', 'Porcentaje Total']]

# Crear un nuevo DataFrame para almacenar la tabla actualizada con porcentajes
tabla_final_relieve = tabla_produccion_subregion_relieve.copy()

# Iterar sobre cada columna de relieve para calcular y añadir columnas de porcentaje
for relieve in relieve_cols:
    # Obtener la producción total para el relieve actual
    total_relieve_produccion = tabla_produccion_subregion_relieve.loc['Total General', relieve]

    # Calcular porcentajes, manejando la división por cero para relieves sin producción
    if total_relieve_produccion > 0:
        tabla_final_relieve[f'{relieve} %'] = (tabla_produccion_subregion_relieve[relieve] / total_relieve_produccion) * 100
    else:
        tabla_final_relieve[f'{relieve} %'] = 0.0

    # Formatear la columna de porcentaje
    tabla_final_relieve[f'{relieve} %'] = tabla_final_relieve[f'{relieve} %'].map('{:,.2f}%'.format)

# Reordenar las columnas para intercalar producción original y nuevas columnas de porcentaje
ordered_columns_relieve = []
for relieve in relieve_cols:
    ordered_columns_relieve.append(relieve)
    ordered_columns_relieve.append(f'{relieve} %')
ordered_columns_relieve.append('Total General')
ordered_columns_relieve.append('Porcentaje Total')

# Aplicar el nuevo orden de columnas
tabla_final_relieve = tabla_final_relieve[ordered_columns_relieve]

# Crear la fila 'Subtotal' con los totales y porcentajes correctos
subtotal_values_relieve = []
for col in tabla_final_relieve.columns:
    if col in relieve_cols: # Esto es una columna de producción por relieve (ej., 'Costero')
        subtotal_values_relieve.append(tabla_produccion_subregion_relieve.loc['Total General', col])
    elif col.endswith(' %') and col.replace(' %', '') in relieve_cols: # Esto es una columna de porcentaje por relieve (ej., 'Costero %')
        relieve_name = col.replace(' %', '')
        relieve_total_production = tabla_produccion_subregion_relieve.loc['Total General', relieve_name]
        if grand_total_production_relieve > 0:
            percentage_of_valley_total_relieve = (relieve_total_production / grand_total_production_relieve) * 100
        else:
            percentage_of_valley_total_relieve = 0.0
        subtotal_values_relieve.append('{:,.2f}%'.format(percentage_of_valley_total_relieve))
    elif col == 'Total General':
        subtotal_values_relieve.append(grand_total_production_relieve)
    elif col == 'Porcentaje Total':
        subtotal_values_relieve.append('100.00%')

# Crear una Serie con los valores del subtotal, y establecer su nombre como la etiqueta del índice
subtotal_row_series_relieve = pd.Series(subtotal_values_relieve, index=tabla_final_relieve.columns, name='Subtotal')

# Concatenar el DataFrame existente con la nueva fila de subtotal
tabla_final_relieve = pd.concat([tabla_final_relieve, subtotal_row_series_relieve.to_frame().T])

# Quitar la notación científica y mostrar hasta 2 decimales para otras columnas float
pd.options.display.float_format = '{:,.2f}'.format
display(tabla_final_relieve)
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import unicodedata
from google.colab import files # Import files for upload/download

# Lista definitiva para una visualización limpia y con identidad regional
seleccion_estrategica = [
    'Plátano', 'Piña', 'Banano', 'Aguacate', 'Vid', 'Chontaduro',
    'Papaya',]

# 1. Función de normalización para asegurar el match de municipios
def norm(texto):
    if not isinstance(texto, str): return texto
    return ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn').strip().upper()

# --- PREPARACIÓN DEL DATAFRAME DE PRODUCCIÓN (df) PARA TODOS LOS AÑOS ---
# Usamos 'df' que ya está cargado y preprocesado en el notebook
df_prod_base = df.copy() # Latitud y Longitud ya convertidas a numérico en el paso de limpieza inicial

# NO SE FILTRA POR AÑO, se considera toda la serie temporal.

print('\n--- DEBUG: df_prod_base antes de filtrar cultivos y dropna (Lat/Lon ya convertidos) ---')
display(df_prod_base[['Cultivo', 'Latitud', 'Longitud']].head())
print(f'Valores NaN en Latitud en df_prod_base: {df_prod_base['Latitud'].isnull().sum()}')
print(f'Valores NaN en Longitud en df_prod_base: {df_prod_base['Longitud'].isnull().sum()}')

# 3. El DF Final es ahora df_prod_base
df_final = df_prod_base.copy()

# 4. Filtrar cultivos de interés y aplicar Jitter (Dispersión)
# Ahora usaremos los cultivos definidos en 'seleccion_estrategica'
cultivos_interes = seleccion_estrategica
df_mapa_detalle = df_final[df_final['Cultivo'].isin(cultivos_interes)].copy()

# Agrupar por Municipio, Cultivo, Latitud, Longitud y sumar la Produccion_toneladas para el acumulado total
df_mapa_detalle = df_mapa_detalle.groupby(['Municipio', 'Cultivo', 'Latitud', 'Longitud'])['Produccion_toneladas'].sum().reset_index()

print('\n--- DEBUG: df_mapa_detalle antes de dropna ---')
display(df_mapa_detalle[['Cultivo', 'Latitud', 'Longitud', 'Produccion_toneladas']].head())
print(f'Valores NaN en Latitud en df_mapa_detalle antes de dropna: {df_mapa_detalle['Latitud'].isnull().sum()}')
print(f'Valores NaN en Longitud en df_mapa_detalle antes de dropna: {df_mapa_detalle['Longitud'].isnull().sum()}')

df_mapa_detalle = df_mapa_detalle.dropna(subset=['Latitud', 'Longitud']) # Quitar los que no tienen coordenadas válidas

# Verificar si hay datos para mapear después de los filtros
if df_mapa_detalle.empty:
    print("No hay datos para los cultivos de interés en los años considerados con coordenadas válidas para generar el mapa.")
else:
    # Ordenar por Produccion_toneladas ascendente para que las burbujas más pequeñas se dibujen al final (arriba)
    df_mapa_detalle = df_mapa_detalle.sort_values(by='Produccion_toneladas', ascending=True).copy()

    # Aumentar ligeramente el jitter para mayor dispersión
    stdev = 0.012  # Aumentado de 0.008 a 0.012
    df_mapa_detalle['lat_j'] = df_mapa_detalle['Latitud'] + np.random.normal(0, stdev, len(df_mapa_detalle))
    df_mapa_detalle['lon_j'] = df_mapa_detalle['Longitud'] + np.random.normal(0, stdev, len(df_mapa_detalle))

    # 5. Paleta de Colores de Alta Visibilidad
    # Colores personalizados según la solicitud del usuario
    color_map_high_viz = {
        'Piña': '#FFD700',       # Gold
        'Banano': '#FFFF00',     # Yellow
        'Aguacate': '#39FF14',   # Fluorescent Green
        'Papaya': '#FF8C00',     # DarkOrange
        'Plátano': '#006400',    # Very Dark Green
        'Vid': '#800080',        # Purple
        'Chontaduro': '#FF0000' # Red
    }

    # Asegurar que todos los cultivos en df_mapa_detalle tengan un color asignado
    # Si un cultivo no está en color_map_high_viz, se le asignará un color de Dark24
    unique_cultivos_en_mapa = df_mapa_detalle['Cultivo'].unique()
    colors_default = px.colors.qualitative.Dark24
    for i, cultivo in enumerate(unique_cultivos_en_mapa):
        if cultivo not in color_map_high_viz:
            color_map_high_viz[cultivo] = colors_default[i % len(colors_default)]

    # 6. Generar el Mapa
    fig = px.scatter_mapbox(
        df_mapa_detalle,
        lat='lat_j', lon='lon_j',
        size='Produccion_toneladas',
        color='Cultivo',
        color_discrete_map=color_map_high_viz,
        size_max=40,
        zoom=7.5,
        center=dict(lat=4.437973, lon=-76.880381), # Actualizado con las coordenadas del usuario
        mapbox_style="carto-positron",
        title='Mapa de Producción Acumulada por Cultivo (Selección Estratégica) - Todos los Años',
        hover_name='Municipio',
        opacity=0.8
    )

    # Calcular la producción total del valle para los porcentajes de la leyenda
    total_produccion_valle = df['Produccion_toneladas'].sum()

    # Preparar los elementos de la leyenda
    legend_items = []

    # Calcular el porcentaje para cada cultivo estratégico
    for cultivo in seleccion_estrategica:
        cultivo_prod = df[df['Cultivo'] == cultivo]['Produccion_toneladas'].sum()
        cultivo_percent = (cultivo_prod / total_produccion_valle) * 100
        legend_items.append(f"{cultivo}: {cultivo_percent:.2f}%")

    # Calcular la producción para los cultivos no estratégicos ("Demas")
    non_strategic_crops_prod = df[~df['Cultivo'].isin(seleccion_estrategica)]['Produccion_toneladas'].sum()
    non_strategic_percent = (non_strategic_crops_prod / total_produccion_valle) * 100
    legend_items.append(f"Demas: {non_strategic_percent:.2f}%")

    legend_text = "<br>".join(legend_items)

    fig.update_layout(
        margin={"r":0,"t":50,"l":0,"b":0},
        annotations=[
            dict(
                text=f"<b>Porcentaje de Producción Total Valle:</b><br>{legend_text}",
                showarrow=False,
                xref="paper", yref="paper",
                x=0.01, y=0.99, # Posición en la esquina superior izquierda
                xanchor="left", yanchor="top",
                align="left",
                bgcolor="rgba(255,255,255,0.7)", # Fondo semi-transparente
                bordercolor="black",
                borderwidth=1,
                font=dict(size=10)
            )
        ]
    )
    fig.show()

    # Guardar y descargar el mapa HTML
    html_file_name = "mapa_produccion_seleccion_estrategica_acumulada_total.html"
    fig.write_html(html_file_name)

    print(f"Mapa guardado como '{html_file_name}' y listo para descargar.")
    files.download(html_file_name)
```

```python
print('--- Producción de Cultivo de Plátano por Año y Subregión ---')

# Filtrar el DataFrame para el cultivo de 'Plátano'
df_platano = df[df['Cultivo'] == 'Plátano'].copy()

# Calcular la producción total anual por subregión
produccion_platano_anual = df_platano.groupby(['año', 'Subregion'])['Produccion_toneladas'].sum().unstack(fill_value=0)

# Calcular el total anual de producción de plátano
produccion_platano_anual['Total Anual'] = produccion_platano_anual.sum(axis=1)

# Calcular el porcentaje de cada subregión respecto al total anual
subregions = [col for col in produccion_platano_anual.columns if col != 'Total Anual']
for sub in subregions:
    produccion_platano_anual[f'{sub} %'] = (produccion_platano_anual[sub] / produccion_platano_anual['Total Anual']) * 100

# Reordenar las columnas para mostrar Producción y Porcentaje juntos
ordered_cols = []
for sub in subregions:
    ordered_cols.append(sub)
    ordered_cols.append(f'{sub} %')
ordered_cols.append('Total Anual')

produccion_platano_anual = produccion_platano_anual[ordered_cols]

# Añadir una fila de totales generales al final
total_general_platano = pd.DataFrame(produccion_platano_anual.sum()).T
total_general_platano.index = ['Total General']

# Calcular los porcentajes totales generales correctamente
for sub in subregions:
    total_produccion_subregion = df_platano[df_platano['Subregion'] == sub]['Produccion_toneladas'].sum()
    total_produccion_general_cultivo = df_platano['Produccion_toneladas'].sum()
    if total_produccion_general_cultivo > 0:
        total_general_platano[f'{sub} %'] = (total_produccion_subregion / total_produccion_general_cultivo) * 100
    else:
        total_general_platano[f'{sub} %'] = 0.0

# Concatenar el total general al DataFrame principal
produccion_platano_anual = pd.concat([produccion_platano_anual, total_general_platano])

# Formatear las columnas de porcentaje
pd.options.display.float_format = '{:,.2f}'.format
for col in produccion_platano_anual.columns:
    if '%' in col:
        produccion_platano_anual[col] = produccion_platano_anual[col].map('{:,.2f}%'.format)


display(produccion_platano_anual)
```

El modelo de regresión lineal para el Plátano confirma una tendencia de crecimiento sostenido y robusto.
```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

print('--- Modelo de Regresión Lineal para la Producción de Plátano por Año ---')

# Filtrar el DataFrame para el cultivo de 'Plátano'
df_platano_linear = df[df['Cultivo'] == 'Plátano'].copy()

# Asegurarse de que 'año' sea numérico para el modelo
df_platano_linear['año_numerico'] = pd.to_numeric(df_platano_linear['año'])

# Agrupar por año y sumar la producción de plátano
produccion_anual_platano_linear = df_platano_linear.groupby('año_numerico')['Produccion_toneladas'].sum().reset_index()

# Preparar los datos para la regresión lineal
X_linear = produccion_anual_platano_linear[['año_numerico']]
y_linear = produccion_anual_platano_linear['Produccion_toneladas']

# Crear y entrenar el modelo de regresión lineal
model_linear = LinearRegression()
model_linear.fit(X_linear, y_linear)

# Realizar predicciones
y_pred_linear = model_linear.predict(X_linear)

# Coeficiente R-cuadrado para el modelo lineal
r_squared_linear = model_linear.score(X_linear, y_linear)

# Calcular métricas adicionales
mse = mean_squared_error(y_linear, y_pred_linear)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_linear, y_pred_linear)

# Calcular MAPE (Mean Absolute Percentage Error)
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape = mean_absolute_percentage_error(y_linear, y_pred_linear)

# Visualizar los resultados
plt.figure(figsize=(12, 6))
plt.scatter(X_linear, y_linear, color='blue', label='Datos reales de producción')
plt.plot(X_linear, y_pred_linear, color='red', label='Línea de Regresión Lineal')

plt.title('Producción de Plátano por Año con Regresión Lineal')
plt.xlabel('Año')
plt.ylabel('Producción (toneladas)')
plt.grid(True)
plt.legend()
plt.show()

print(f"\nMétricas del Modelo Lineal:")
print(f"- R-cuadrado (R²): {r_squared_linear:.2f}")
print(f"- Error Cuadrático Medio (MSE): {mse:,.2f}")
print(f"- Raíz del Error Cuadrático Medio (RMSE): {rmse:,.2f}")
print(f"- Error Absoluto Medio (MAE): {mae:,.2f}")
print(f"- Error Porcentual Absoluto Medio (MAPE): {mape:,.2f}%")

# Interpretación del R-cuadrado
if r_squared_linear > 0.7:
    print("Interpretación: El modelo lineal explica una gran parte de la variabilidad en la producción anual de plátano. El año es un buen predictor de la producción.")
elif r_squared_linear > 0.4:
    print("Interpretación: El modelo lineal explica una parte moderada de la variabilidad en la producción anual de plátano. El año tiene una influencia notable pero no dominante.")
else:
    print("Interpretación: El modelo lineal explica una pequeña parte de la variabilidad en la producción anual de plátano. El año no es un fuerte predictor de la producción por sí solo, y otros factores pueden ser más influyentes.")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

print('--- Modelo de Regresión Logarítmica para la Producción de Plátano por Año ---')

# Filtrar el DataFrame para el cultivo de 'Plátano' y agrupar la producción anual
df_platano_log = df[df['Cultivo'] == 'Plátano'].copy()
df_platano_log['año_numerico'] = pd.to_numeric(df_platano_log['año'])
produccion_anual_platano_log = df_platano_log.groupby('año_numerico')['Produccion_toneladas'].sum().reset_index()

# Preparar los datos para la regresión logarítmica
# X será el año (variable independiente)
X_log = produccion_anual_platano_log[['año_numerico']]
# y será el logaritmo de la producción (variable dependiente)
# Se añade 1 para evitar log(0) si hubiera producciones de 0 toneladas, aunque en este dataset parece no haber.
# Pero es una buena práctica de robustez.
y_log_transformed = np.log(produccion_anual_platano_log['Produccion_toneladas'] + 1)
y_original = produccion_anual_platano_log['Produccion_toneladas']

# Crear y entrenar el modelo de regresión lineal con la variable dependiente transformada
model_log = LinearRegression()
model_log.fit(X_log, y_log_transformed)

# Realizar predicciones sobre la variable log-transformada
y_pred_log_transformed = model_log.predict(X_log)

# Convertir las predicciones de vuelta a la escala original para la visualización y métricas
y_pred_original_scale = np.exp(y_pred_log_transformed) - 1 # Restar 1 si se sumó antes del log

# Coeficiente R-cuadrado
r_squared_log = r2_score(y_original, y_pred_original_scale)

# Calcular métricas adicionales en la escala original
mse_log = mean_squared_error(y_original, y_pred_original_scale)
rmse_log = np.sqrt(mse_log)
mae_log = mean_absolute_error(y_original, y_pred_original_scale)

# Visualizar los resultados
plt.figure(figsize=(12, 6))
plt.scatter(X_log, y_original, color='blue', label='Datos reales de producción')
plt.plot(X_log, y_pred_original_scale, color='red', label='Línea de Regresión Logarítmica')

plt.title('Producción de Plátano por Año con Regresión Logarítmica')
plt.xlabel('Año')
plt.ylabel('Producción (toneladas)')
plt.grid(True)
plt.legend()
plt.show()

print(f"\nMétricas del Modelo Logarítmico:")
print(f"- R-cuadrado (R²): {r_squared_log:.2f}")
print(f"- Error Cuadrático Medio (MSE): {mse_log:,.2f}")
print(f"- Raíz del Error Cuadrático Medio (RMSE): {rmse_log:,.2f}")
print(f"- Error Absoluto Medio (MAE): {mae_log:,.2f}")

# Interpretación del R-cuadrado
if r_squared_log > 0.7:
    print("Interpretación: El modelo logarítmico explica una gran parte de la variabilidad en la producción anual de plátano. La transformación logarítmica ha ayudado a capturar la tendencia de crecimiento.")
elif r_squared_log > 0.4:
    print("Interpretación: El modelo logarítmico explica una parte moderada de la variabilidad en la producción anual de plátano. Otros factores pueden ser más influyentes o se podría considerar otra forma de regresión.")
else:
    print("Interpretación: El modelo logarítmico explica una pequeña parte de la variabilidad en la producción anual de plátano. El año no es un fuerte predictor de la producción por sí solo o la transformación logarítmica no es la más adecuada.")
```

**19. Modelo de suavizado exponencial:**

Es una opción más robusta para pronosticar series de tiempo, especialmente para proyecciones a largo plazo donde la regresión polinómica puede generar valores poco realistas.

Este modelo logra explicar el 77% de la variabilidad en la producción histórica de plátano, lo cual es un buen ajuste. Comparado con el modelo polinómico de grado 3 (R² = 0.81), el suavizado exponencial tiene un R² ligeramente menor, pero a menudo ofrece proyecciones más estables y realistas a largo plazo, ya que no tiende a generar valores negativos como los modelos polinómicos.

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

print('--- Modelo de Suavizado Exponencial para la Producción de Plátano por Año ---')

# Filtrar el DataFrame para el cultivo de 'Plátano'
df_platano_es = df[df['Cultivo'] == 'Plátano'].copy()

# Asegurarse de que 'año' sea numérico para agrupar
df_platano_es['año_numerico'] = pd.to_numeric(df_platano_es['año'])

# Agrupar por año y sumar la producción de plátano
produccion_anual_platano_es = df_platano_es.groupby('año_numerico')['Produccion_toneladas'].sum().reset_index()

# Convertir 'año_numerico' a formato de fecha y establecerlo como índice para la serie de tiempo
produccion_anual_platano_es['año_numerico'] = pd.to_datetime(produccion_anual_platano_es['año_numerico'], format='%Y')
produccion_anual_platano_es = produccion_anual_platano_es.set_index('año_numerico')
produccion_anual_platano_es.index.freq = 'YS' # Establecer frecuencia anual (Year Start)

# Entrenar el modelo de Suavizado Exponencial
# Se usa tendencia aditiva ('add') y sin estacionalidad ('seasonal=None') para datos anuales
model_es = ExponentialSmoothing(
    produccion_anual_platano_es['Produccion_toneladas'],
    trend='add',
    seasonal=None,
    initialization_method="estimated"
).fit()

# Realizar predicciones para los años futuros (22 años)
forecast_steps = 22
forecast_es = model_es.forecast(steps=forecast_steps)

# Asegurar que no haya valores negativos en el pronóstico
forecast_es = forecast_es.apply(lambda x: max(0, x))

# Calcular métricas de rendimiento para los datos históricos
y_true = produccion_anual_platano_es['Produccion_toneladas']
y_pred_fitted = model_es.fittedvalues

r2_es = r2_score(y_true, y_pred_fitted)
mse_es = mean_squared_error(y_true, y_pred_fitted)
rmse_es = np.sqrt(mse_es)
mae_es = mean_absolute_error(y_true, y_pred_fitted)

# Preparar DataFrame para las proyecciones futuras
df_predictions_es = pd.DataFrame({
    'año_numerico': pd.to_numeric(forecast_es.index.year),
    'Produccion_toneladas': forecast_es.values
})

# Concatenar los datos históricos con las predicciones para la visualización
combined_df_es = pd.concat([
    produccion_anual_platano_es.reset_index().rename(columns={'año_numerico': 'año'}),
    df_predictions_es.rename(columns={'año_numerico': 'año'})
])

# Visualizar los resultados
plt.figure(figsize=(14, 7))
plt.scatter(produccion_anual_platano_es.index.year, produccion_anual_platano_es['Produccion_toneladas'], color='blue', label='Datos reales históricos')
plt.plot(produccion_anual_platano_es.index.year, model_es.fittedvalues, color='green', label='Ajuste del Modelo (Suavizado Exponencial)')
plt.plot(df_predictions_es['año_numerico'], df_predictions_es['Produccion_toneladas'], color='red', linestyle='--', label='Proyección Futura (hasta 2044)')

plt.title('Producción de Plátano Histórica y Proyectada (hasta 2044) con Suavizado Exponencial')
plt.xlabel('Año')
plt.ylabel('Producción (toneladas)')
plt.grid(True)
plt.legend()
plt.show()

print(f"\nMétricas del Modelo de Suavizado Exponencial:")
print(f"- R-cuadrado (R²): {r2_es:.2f}")
print(f"- Error Cuadrático Medio (MSE): {mse_es:,.2f}")
print(f"- Raíz del Error Cuadrático Medio (RMSE): {rmse_es:,.2f}")
print(f"- Error Absoluto Medio (MAE): {mae_es:,.2f}")

# Interpretación del R-cuadrado
if r2_es > 0.7:
    print("Interpretación: El modelo de suavizado exponencial explica una gran parte de la variabilidad en la producción anual de plátano, ofreciendo un buen ajuste y proyecciones estables.")
elif r2_es > 0.4:
    print("Interpretación: El modelo de suavizado exponencial explica una parte moderada de la variabilidad en la producción anual de plátano. Podría haber otros factores influyentes.")
else:
    print("Interpretación: El modelo de suavizado exponencial explica una pequeña parte de la variabilidad en la producción anual de plátano. Es posible que no sea el modelo más adecuado para esta serie de tiempo.")

print('\nProyecciones de Producción de Plátano por Año (hasta 2044):')
df_predictions_es['Produccion_toneladas'] = df_predictions_es['Produccion_toneladas'].apply(lambda x: f'{x:,.2f}')
display(df_predictions_es.set_index('año_numerico'))
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import math # Added import for math.sin

print('--- Creando Dataset ProyeccionFrutales2044 ---')

# Re-generate the ExponentialSmoothing model forecast (forecast_es)
# as the original cell generating it was modified.
# Filter the DataFrame for the cultivation of 'Plátano'
df_platano_es = df[df['Cultivo'] == 'Plátano'].copy()

# Ensure 'year' is numeric to group
df_platano_es['año_numerico'] = pd.to_numeric(df_platano_es['año'])

# Group by year and sum banana production
produccion_anual_platano_es = df_platano_es.groupby('año_numerico')['Produccion_toneladas'].sum().reset_index()

# Set the year as index for the time series
# Convert 'año_numerico' to date format and set it as index
produccion_anual_platano_es['año_numerico'] = pd.to_datetime(produccion_anual_platano_es['año_numerico'], format='%Y')
produccion_anual_platano_es = produccion_anual_platano_es.set_index('año_numerico')
produccion_anual_platano_es.index.freq = 'YS' # Set annual frequency (Year Start)

# Train the Exponential Smoothing model
model_es = ExponentialSmoothing(
    produccion_anual_platano_es['Produccion_toneladas'],
    trend='add',
    seasonal=None, # No seasonality for annual data usually
    initialization_method="estimated"
).fit()

# Years to project (22 years from the last year of the dataset, i.e., up to 2044)
forecast_steps = 22 # Project 22 years from the last year (2022)

# Make predictions
forecast_es = model_es.forecast(steps=forecast_steps)

# Ensure no negative values (despite additive trend, it could happen if trend is very negative)
forecast_es = forecast_es.apply(lambda x: max(0, x))

# --- Begin User Requested Projection Logic ---

# 1. Filter original data for 'Plátano' in 2022
df_platano_2022 = df[(df['Cultivo'] == 'Plátano') & (df['año'] == '2022')].copy()

# Explicitly cast 'Temp, Promedio (°C)' and 'Precipitación (mm/año)' to float
# to prevent FutureWarning when assigning float values later.
df_platano_2022['Temp, Promedio (°C)'] = df_platano_2022['Temp, Promedio (°C)'].astype(float)
df_platano_2022['Precipitación (mm/año)'] = df_platano_2022['Precipitación (mm/año)'].astype(float)

# Store original values for projection, ensuring they are numeric
# These will be the initial values for the year-over-year projection
current_production_per_record = pd.to_numeric(df_platano_2022['Produccion_toneladas'], errors='coerce')
current_temp_per_record = pd.to_numeric(df_platano_2022['Temp, Promedio (°C)'], errors='coerce')
current_precip_per_record = pd.to_numeric(df_platano_2022['Precipitación (mm/año)'], errors='coerce')
original_relieve_2022 = df_platano_2022['Relieve Predominante'].copy() # Relief doesn't change

# Drop the columns that will be projected or are not needed in the base copy
# This df_base_for_projection now holds the static info for each record (municipio, cultivo, lat, lon, etc.)
df_base_for_projection = df_platano_2022.drop(columns=[
    'Produccion_toneladas',
    'Temp, Promedio (°C)',
    'Precipitación (mm/año)',
    'año'
])

# Define projection parameters
start_year = 2023
end_year = 2044
projection_years = range(start_year, end_year + 1)

# Define growth/decrease factors based on 'Relieve Predominante'
production_growth_factors = {
    'Costero': 0.002,   # Crecimiento leve
    'Montañoso': 0.01,  # Crecimiento mayor (el calor ayuda en altura)
    'Plano': -0.015     # Caída por estrés térmico y falta de agua
}

temp_increase_per_year = {
    'Costero': 0.03,   # +0.6°C en 20 años
    'Montañoso': 0.04, # +0.8°C en 20 años
    'Plano': 0.06      # +1.2°C en 20 años (Zona más crítica)
}

precip_growth_rates = {
    'Costero': -0.001,   # Disminución leve (-0.1% anual)
    'Montañoso': 0.005,  # Aumento moderado (+0.5% anual)
    'Plano': 0.003       # Aumento leve (+0.3% anual)
}

all_projected_dfs = []

# Loop through each projection year
for year in projection_years:
    # Create a DataFrame for the current projected year, keeping static info
    df_current_year = df_base_for_projection.copy()
    df_current_year['año'] = str(year)

    # Initialize lists to hold projected values for the current year
    projected_productions_for_year = []
    projected_temps_for_year = []
    projected_precips_for_year = []

    # Iterate over each record (original index and its corresponding relief)
    for idx, relief in original_relieve_2022.items():
        # Apply the user's requested logic for Temperature
        # current_temp_per_record.loc[idx] is the temperature from the previous year for this record
        current_temp_per_record.loc[idx] += temp_increase_per_year.get(relief, 0) + np.random.normal(0, 0.05)
        projected_temps_for_year.append(current_temp_per_record.loc[idx])

        # Apply the user's requested logic for Precipitation
        # current_precip_per_record.loc[idx] is the precipitation from the previous year for this record
        variacion_ciclica = 0.05 * math.sin(year)
        current_precip_per_record.loc[idx] *= (1 + precip_growth_rates.get(relief, 0) + variacion_ciclica)
        projected_precips_for_year.append(current_precip_per_record.loc[idx])

        # Apply the user's requested logic for Production
        # current_production_per_record.loc[idx] is the production from the previous year for this record
        current_production_per_record.loc[idx] *= (1 + production_growth_factors.get(relief, 0))
        projected_productions_for_year.append(current_production_per_record.loc[idx])

    # Assign the calculated projected values for the current year to the DataFrame
    df_current_year['Produccion_toneladas'] = projected_productions_for_year
    df_current_year['Temp, Promedio (°C)'] = projected_temps_for_year
    df_current_year['Precipitación (mm/año)'] = projected_precips_for_year

    all_projected_dfs.append(df_current_year)

# Concatenate all yearly DataFrames
ProyeccionFrutales2044 = pd.concat(all_projected_dfs, ignore_index=True)

print(f"Dataset 'ProyeccionFrutales2044' creado con {len(ProyeccionFrutales2044)} registros.")
print("\n--- Primeras filas del Dataset ProyeccionFrutales2044 ---")
display(ProyeccionFrutales2044.head())

print("\n--- Información del Dataset ProyeccionFrutales2044 (df.info()) ---")
ProyeccionFrutales2044.info()

print("\n--- Estadísticas Descriptivas del Dataset ProyeccionFrutales2044 (df.describe()) ---")
display(ProyeccionFrutales2044.describe())
```
